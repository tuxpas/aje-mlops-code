# 3. Combinamos PS de los paises

```python
# Nombrar paises
paises_a_concatenar = [ps_pan, ps_ecuador, ps_costarica, ps_mexico, ps_guatemala, ps_nicaragua, ps_peru] 
```

```python
final_ps = pd.concat(paises_a_concatenar, ignore_index=True)
```

```python
final_ps.dtypes
```

```python
final_ps["Compania"] = final_ps["Compania"].apply(lambda x: str(int(x)).rjust(4, "0"))
final_ps["Sucursal"] = final_ps["Sucursal"].apply(lambda x: str(int(x)).rjust(2, "0"))
```

```python
final_ps['tipoRecomendacion'] = final_ps.groupby(['Pais', 'Compania', 'Sucursal', 'Cliente'])\
               .cumcount()\
               .apply(lambda x: f'PS{x+1}')
```

```python
final_ps["ultFecha"] = ''
final_ps["Destacar"] = 'true'
```

## Concatenar con Pedido Estrategico y Recurrente (ECUADOR) -> (NO CRUZA, CONCATENA)

---

```python
# Leemos PEDIDO ESTRATEGICO
s3 = boto3.client('s3')
bucket_name = "aje-analytics-ps-backup"
file_key = f"Pedido_Estrategico/Ecuador/Output/estr_base_pedidos_{fecha_tomorrow}.csv"

# 2. Descargar el objeto del S3 a la memoria RAM
response = s3.get_object(Bucket=bucket_name, Key=file_key)

# 3. Leerlo con Pandas sin pasar por s3fs
p_est = pd.read_csv(io.BytesIO(response['Body'].read()), 
    dtype={'Compania': str, 'Sucursal': str , 'Producto': str})

p_est["ultFecha"] = ''
p_est["Destacar"] = 'true'
```

```python
p_est
```

```python
# Leemos PEDIDO RECURRENTE
s3 = boto3.client('s3')
bucket_name = "aje-analytics-ps-backup"
file_key = f"Pedido_Recurrente/Ecuador/Output/recu_base_pedidos_{fecha_tomorrow}.csv"

# 2. Descargar el objeto del S3 a la memoria RAM
response = s3.get_object(Bucket=bucket_name, Key=file_key)

# 3. Leerlo con Pandas sin pasar por s3fs
p_rec = pd.read_csv(io.BytesIO(response['Body'].read()), 
    dtype={'Compania': str, 'Sucursal': str , 'Producto': str})

p_rec["Destacar"] = 'true'
```

```python
p_rec
```

```python
final_ps = pd.concat([p_est, p_rec, final_ps], ignore_index=True)
```
